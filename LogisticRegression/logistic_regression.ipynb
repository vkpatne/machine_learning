## Logistic regession

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import LabelEncoder

import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = fetch_openml(name='diabetes', version=1, as_frame=False)
## independent features
X=pd.DataFrame(df.data, columns=df.feature_names)
X.head()
```

```python
## Dependent feature
y=pd.DataFrame(df.target, columns=['target'])

# 1. Convert string labels to numeric
le = LabelEncoder()
y_numeric = le.fit_transform(y)  # Now contains 0/1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_numeric, test_size=0.33, random_state=42)
```

```python
params = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'max_iter': [100, 200, 300]}
```

```python
model1=LogisticRegression(C=100, max_iter=100)
```

```python
from sklearn.model_selection import GridSearchCV
model=GridSearchCV(model1, params, cv=5, scoring='f1')
```

```python
model.fit(X_train, y_train)
```

```python
model.best_score_
```

```python
model.best_params_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score = model.score(X_test, y_test)
accuracy_score
```

```python
print(classification_report(y_test, y_pred))
```
